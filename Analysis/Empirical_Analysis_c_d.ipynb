```python
import pandas as pd
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

## Checking how the number of visited nodes, i.e., the number of items pushed into the queue, changes as:
### k*v+h where k<h
### h*v where k>=h

```python
files={200:"multisource_bfs_200nodes_6034edges.pkl" ,300:"multisource_bfs_300nodes_13381edges.pkl", 400:"multisource_bfs_400nodes_LotsOfedges.pkl"}
for key, values in files.items():
    data=pd.read_pickle(values)
    print(data)
    hospitals=[]
    visited_nodes=[]
    for element in data:
        hospitals.append(element[0])
        visited_nodes.append(element[1])
    title="No. of Hospital Nodes vs No. of Nodes algorithm visits where k="+str(data[0][2])+" For "+str(key)+" nodes "
    plt.plot(hospitals,visited_nodes)
    plt.title(title)
    plt.xlabel('Hospitals')
    plt.ylabel('Visited Nodes')
    plt.show()
#empirical analysis done by varying the number of nodes, keeping k=10 and probability of edges occurring = 0.3
    
```

## Checking how time elapsed changes with the number of edges
### It is a linear function of the number of edges

```python
files={200:"SparseVsDenseGraph.pkl" }
for key, values in files.items():
    data=pd.read_pickle(values)
    print(data)
    hospitals=[]
    visited_nodes=[]
    for element in data:
        hospitals.append(element[0])
        visited_nodes.append(element[1])
    
    plt.plot(hospitals,visited_nodes)
    plt.title("Variation of time elapsed with number of edges")
    plt.xlabel('Number of edges')
    plt.ylabel('Time elapsed')
    plt.show()
#empirical analysis for a graph of 200 nodes, probability varying from 0.1 onwards
#as the graph gets denser, time taken increases
```
